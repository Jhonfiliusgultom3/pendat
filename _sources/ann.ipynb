# MLP CLASSIFIER 


atau (Multi-Layer Perceptron Classifier) adalah salah satu jenis model yang digunakan dalam machine learning untuk masalah klasifikasi. MLP Classifier adalah bentuk yang lebih umum dari Jaringan Saraf Tiruan (Artificial Neural Network) dengan beberapa lapisan tersembunyi (hidden layers) yang menghubungkan antara input dan output.

MLP Classifier menggunakan algoritma pembelajaran yang disebut backpropagation untuk melatih modelnya. Proses pelatihan melibatkan iterasi berulang dalam menghitung dan memperbarui bobot dan bias di setiap lapisan hingga model dapat mempelajari pola yang tepat dalam data.

#### ***Import Librari***

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
#Membaca data Csv berbentuk url
data_iris = pd.read_csv('https://raw.githubusercontent.com/Jhonfiliusgultom3/data/main/iris.csv')
```

```python
data_iris.head(20)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
y = data_iris['variety']
x = data_iris.drop(['variety'], axis=1)
```

#### ***Membagi data menjadi data latih dan uji***

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=10)
```

```python
#untuk inisialisasi objek MLP Classifier dengan beberapa parameter yang ditentukan
clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=1000, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.001)
```

#### ***Menghitung akurasi***

```python
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(y_test, y_pred)
```

```python
import pickle #mengimpor modul pickle, yang menyediakan fungsionalitas untuk menyimpan dan memuat objek Python.
with open('mlpku','wb') as f: #konstruksi yang digunakan untuk membuka file dengan nama 'mlpku' dalam mode write-binary ('wb')
    pickle.dump(clf,f)
```
