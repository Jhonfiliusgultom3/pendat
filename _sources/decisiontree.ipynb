# IMPLEMENTASI DECISION TREE (DATA IRIS)


Decision Tree (Pohon Keputusan) adalah salah satu algoritma pembelajaran mesin yang digunakan untuk masalah klasifikasi dan regresi. Algoritma ini menggunakan struktur berbentuk pohon yang berisi serangkaian keputusan dan aturan untuk memprediksi label atau nilai target dari data input.

Pohon keputusan terdiri dari simpul (node) yang merepresentasikan keputusan atau atribut, cabang (branch) yang merepresentasikan hasil keputusan, dan daun (leaf) yang merepresentasikan klasifikasi atau prediksi akhir.

Proses pembentukan pohon keputusan dimulai dengan membagi data berdasarkan atribut-atribut yang ada. Pada setiap langkah, atribut yang memiliki kemampuan terbaik untuk memisahkan kelas target akan dipilih sebagai node pengambilan keputusan. Pemisahan dilakukan dengan meminimalkan ketidakmurnian (impurity) pada setiap cabang.

#### ***Import Librari***

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
#Membaca data Csv berbentuk url
data_iris = pd.read_csv('https://raw.githubusercontent.com/Jhonfiliusgultom3/data/main/iris_data.csv')
```

```python
data_iris.head(20)
```

```python
y = data_iris["variety"] #diprediksi (variety) disimpan dalam variabel y. Ini berarti y akan berisi label atau kelas dari data.
X = data_iris.drop(columns=["variety"])# variabel X akan berisi fitur-fitur atau atribut-atribut dari data.
# Perintah drop digunakan untuk menghapus kolom "variety" dari data dan menyimpan hasilnya dalam variabel X

#Pada langkah ini, objek DecisionTreeClassifier diinisialisasi dengan menggunakan parameter criterion="gini"
clf = tree.DecisionTreeClassifier(criterion="gini")

clf = clf.fit(X, y)
```

```python
fig = plt.figure(figsize=(10,10))
_ = tree.plot_tree(clf, feature_names=list(data_iris.columns.values)[:4], class_names=list(data_iris.columns.values)[4] ,filled=True)
# list(data_iris.columns.values)[:4] untuk mengambil nama kolom kelima dari dataset sebagai nama kelas.
# Parameter ini menentukan apakah node pohon akan diisi dengan warna sesuai dengan kelas dominan atau tidak. Jika diatur sebagai True
```

```python
x = data_iris.iloc[:,0:4]
y = data_iris.iloc[:,4]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.3, random_state=10)

x_train
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train,y_train)

hasil_prediksi = model .predict(x_test)


print(accuracy_score(hasil_prediksi,y_test))
```
